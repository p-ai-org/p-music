```python
#convert number of ratings into float
#df['Number of Ratings'] = df['Number of Ratings'].str.replace(',', '').astype('float64')
#df['Number of Ratings'].head()
df = df.drop(df.columns[0], axis = 1)
```

```python
#df['Release Day'] = df['Release Day'].astype('int64')
```

```python
#monthname - monthnumber dictionary
import calendar
df['Release Month'].head()
month = calendar.month_name[1:]
month_num = range(1, 13)
month_dict = dict(zip(month, month_num))
#month_dict
```

```python
clean_month = df['Release Month'].dropna()
clean_month = clean_month.apply(lambda x: month_dict[x])
month_mean = clean_month.mean()
df['Release Month'] = df['Release Month'].apply(lambda x: month_dict[x] if x==x else month_mean)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

heatmap_data = df.select_dtypes(include = "number")
heatmap_data.head()
heatmap = heatmap_data.corr()
heatmap = heatmap.style.background_gradient()
#strongly correlated variables: leave only one and drop other. they're giving same information. 
heatmap
```
